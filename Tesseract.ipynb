# Text extraction using OpenCV from an image with TABLE

```python
# Upload an image file
print("Please upload an image file for OCR:")
uploaded = files.upload()

# Get the uploaded file name
image_path2 = list(uploaded.keys())[0]

# Preprocess the image for better OCR results
def preprocess_image(image_path2):
    # Read image with OpenCV
    image2 = cv2.imread(image_path2, cv2.IMREAD_COLOR)

    # Convert to grayscale
    gray = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)

    # Enhance contrast and apply thresholding
    gray = cv2.equalizeHist(gray)
    _, binary = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

    # Dilation to connect text regions
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (2, 2))
    processed_image = cv2.dilate(binary, kernel, iterations=1)

    # Save processed image for verification
    processed_path = "processed_image.png"
    cv2.imwrite(processed_path, processed_image)
    return processed_path

# Preprocess and save the image
processed_path = preprocess_image(image_path2)

# Open and display the uploaded image
image2 = Image.open(image_path2)
plt.figure(figsize=(8, 6))
plt.imshow(image2)
plt.axis('off')
plt.title("Uploaded Image")
plt.show()

# Perform OCR and extract the text as a single string
extracted_text = pytesseract.image_to_string(Image.open(processed_path))

# Clean up the text (optional)
formatted_text = extracted_text.strip()

# Display the extracted text
print("Extracted Text:")
print(formatted_text)
```

# Text extraction from a PDF

```python
!apt-get install poppler-utils tesseract-ocr
from pdf2image import convert_from_path

# Define a function to process the PDF
def extract_text_from_pdf(pdf_path, output_dir="output_images", dpi=300):
    # Create a directory to save images
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # Convert PDF to images
    print("Converting PDF to images...")
    images = convert_from_path(pdf_path, dpi=dpi, fmt='jpeg', output_folder=output_dir)

    extracted_text = []
    print("Processing images with OCR...")
    for i, image in enumerate(images):
        # Convert PIL Image to OpenCV format
        open_cv_image = cv2.cvtColor(np.array(image), cv2.COLOR_RGB2BGR)

        # Convert to grayscale
        gray = cv2.cvtColor(open_cv_image, cv2.COLOR_BGR2GRAY)

        # Apply adaptive thresholding to clean the image
        processed_image = cv2.adaptiveThreshold(
            gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2
        )

        # Save the processed image for debugging (optional)
        output_image_path = f"{output_dir}/page_{i+1}.jpg"
        cv2.imwrite(output_image_path, processed_image)

        # Display the image
        plt.figure(figsize=(10, 6))
        plt.imshow(cv2.cvtColor(processed_image, cv2.COLOR_BGR2RGB))
        plt.title(f"Processed Image - Page {i+1}")
        plt.axis('off')
        plt.show()

        # OCR to extract text
        text = pytesseract.image_to_string(processed_image, config='--oem 3 --psm 6')
        extracted_text.append(text)
        print(f"Extracted text from page {i+1}.")

    # Combine all text into a single output
    full_text = "\n\n".join(extracted_text)

    # Save text to a file
    with open("Extracted_Text.txt", "w") as text_file:
        text_file.write(full_text)

    print("\nText extraction complete. Results saved to 'Extracted_Text.txt'.")
    return full_text

# Upload a PDF file
print("Please upload your PDF file:")
uploaded = files.upload()

# Get the uploaded PDF file path
pdf_path = list(uploaded.keys())[0]  # Take the first uploaded file

# Extract text from the uploaded PDF
text = extract_text_from_pdf(pdf_path)
print("\nExtracted Text:\n", text)
```
